```python
# Function to do compactness pruning

compact_features=[]
def compactness_prune(w1,w2):
    
    conjunctions=['for','and','nor','but','or','yet','so','.'] # Adding full stop also here because compact words cannot be b/w 
                                                                     # two sentences
    iter_count=0

    for i in range(len(df_reviews.Reviews)):
        
        
        count=0
        token=word_tokenize(df_reviews.Reviews[i])

        if w1 in token and w2 in token:

            min_index=(min(token.index(w2),token.index(w1)))+1
            max_index=max(token.index(w2),token.index(w1))

            if not any(word in token[min_index:max_index] for word in conjunctions):
                count=len(token[min_index:max_index])

                if count<=3:
                    iter_count+=1
                    
    if iter_count>=2:
        compact_features.append([w1,w2])

                    #print(df_reviews.Reviews[i]," ----",w1,"&",w2,"are Compact\n")
```

```python
# List of compact features which met our conditions

compact_features=[]
for features in dual_features:
    compactness_prune(str(features[0]),str(features[1]))
print(compact_features)   
print(len(compact_features))
```

```python
# List of dual_features which does not met the criterion and pruned

pruned_features=[]
for i in dual_features:
     if i not in compact_features:
            pruned_features.append(i)
print(pruned_features)
print(len(pruned_features))
```
